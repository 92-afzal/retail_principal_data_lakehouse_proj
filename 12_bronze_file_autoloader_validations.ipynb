```python
spark.sql("""
SELECT COUNT(*) AS total_rows
FROM dbw_retails.bronze.file_transactions
""").show()
```

```python
spark.sql("""
SELECT *
FROM dbw_retails.bronze.file_transactions
LIMIT 10
""").display()
```

```python
spark.sql("""
DESCRIBE dbw_retails.bronze.file_transactions
""").display()
```

```python
spark.sql("""
SELECT
DATE(_ingest_ts) AS ingest_date,
COUNT(*) AS rows
FROM dbw_retails.bronze.file_transactions
GROUP BY DATE(_ingest_ts)
ORDER BY ingest_date DESC
""").display()
```

```python
spark.sql("""
SELECT
regexp_extract(_source_file, '[^/]+$', 0) AS file_name,
COUNT(*) AS rows
FROM dbw_retails.bronze.file_transactions
GROUP BY file_name
""").display()
```

```python
spark.sql("""
SELECT *
FROM dbw_retails.monitoring.pipeline_metrics
WHERE pipeline = 'bronze_files'
ORDER BY processed_at DESC
""").display()
```

```python
spark.sql("""
SELECT
InvoiceNo,
StockCode,
COUNT(*) AS duplicates
FROM dbw_retails.bronze.file_transactions
GROUP BY InvoiceNo, StockCode
HAVING COUNT(*) > 1
LIMIT 10
""").display()
```

```python
spark.sql("SHOW TABLES IN dbw_retails.gold").display()
```

```python
for table in ["product_performance","country_sales","customer_revenue"]:
    spark.sql(f"""
    SELECT '{table}' AS table_name, COUNT(*) AS rows
    FROM dbw_retails.gold.{table}
    """).display()
```

```python
silver_rev = spark.sql("""
SELECT SUM(revenue) AS silver_revenue
FROM dbw_retails.silver.online_retail_transactions
""")

gold_rev = spark.sql("""
SELECT SUM(product_revenue) AS gold_revenue
FROM dbw_retails.gold.product_performance
""")

silver_rev.join(gold_rev).display()
```

```python
spark.sql("""
SELECT *
FROM dbw_retails.gold.customer_revenue
WHERE customer_id IS NULL
""").display()
```

```python
spark.sql("""
SELECT customer_id, COUNT(*)
FROM dbw_retails.gold.customer_revenue
GROUP BY customer_id
HAVING COUNT(*) > 1
""").display()
```

```python
spark.sql("""
SELECT *
FROM dbw_retails.gold.product_performance
ORDER BY product_revenue DESC
LIMIT 10
""").display()
```

Logging checks

```python
spark.sql("""
          SELECT *
FROM dbw_retails.monitoring.pipeline_metrics
ORDER BY processed_at DESC;""").display()
```

```python
spark.sql("DESCRIBE HISTORY dbw_retails.gold.product_performance;").display()
```

```python
spark.sql("""
          SELECT
pipeline,
DATE(processed_at) AS run_date,
SUM(row_count) AS processed_rows
FROM dbw_retails.monitoring.pipeline_metrics
GROUP BY pipeline, run_date
ORDER BY run_date DESC; 
""").display()
```

```python
spark.sql("""SELECT
pipeline,
COUNT(*) AS runs
FROM dbw_retails.monitoring.pipeline_metrics
GROUP BY pipeline;
""").display()
```

```python
spark.sql("""
          SELECT
DATE(processed_at) AS run_date,
SUM(row_count) AS rows_processed
FROM dbw_retails.monitoring.pipeline_metrics
GROUP BY run_date
ORDER BY run_date DESC;
          """).display()
```

```python
spark.sql("""
SELECT
country,
SUM(total_revenue) AS revenue
FROM dbw_retails.gold.country_sales
GROUP BY country
ORDER BY revenue DESC;
""").display()
```
